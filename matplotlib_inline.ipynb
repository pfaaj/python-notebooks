```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt((X**2 + Y**2)/50) #draw a cool ice cream cone, well not really

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.set_zlim(.5, 2.01)

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt

def draw_graph(graph):
    # create networkx graph
    G=nx.Graph()

    # add edges
    for edge in graph:
        G.add_edge(edge[0], edge[1])

    # There are graph layouts like shell, spring, spectral and random.
    # Shell layout usually looks better, so we're choosing it.
    # I will show some examples later of other layouts
    graph_pos = nx.shell_layout(G)

    # draw nodes, edges and labels
    nx.draw_networkx_nodes(G, graph_pos, node_size=1000, node_color='pink', alpha=0.3)
    nx.draw_networkx_edges(G, graph_pos)
    nx.draw_networkx_labels(G, graph_pos, font_size=12, font_family='sans-serif')

    # show graph
    plt.show()

# draw example
# graph is a list of tuples of nodes. Each tuple defining the
# connection between 2 nodes
graph = [(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (4, 6), (5,6)]
```

```python
draw_graph (graph)
```

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig1 = plt.figure()
ax1 = fig1.add_subplot(111, aspect='equal')
w1 = 0.5
h1 = 0.1
x1 = 0.1
y1 = 0.1

class hanoi_piece():
    def __init__(self, w , h, x , y):
        self.w = w
        self.h = h
        self.x = x
        self.y = y
        
def rect(w,h,x = 0.1, y = 0.1):
    ax1.add_patch(
    patches.Rectangle((x, y), w, h,))


def hanoi_init():
    rect(w = w1, h = h1, x = x1, y = y1)
    return hanoi_piece(w1,h1,x1,y1)

def hanoi_add():
    global w1
    global h1
    global y1
    global x1
    w2 = 0.7 * w1
    h2 = h1
    y2 = y1 + h1
    x2 = x1 + (w1/2.0) - (w2/2.0)
    rect(w2, h1, x = x2, y = y2)
    w1 = w2
    h1 = h2
    x1 = x2
    y1 = y2
    return hanoi_piece(w2,h2,x2,y2)
def make_hanoi(num_pieces):
    pieces = []
    pieces.append(hanoi_init())
    for i in range(num_pieces - 1):
        pieces.append(hanoi_add())
    return pieces

pieces_lst = make_hanoi(7)
plt.show()
```
